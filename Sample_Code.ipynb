# 1. Loading the dataset <a name='1.-loading-the-dataset'></a>

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data as pandas dataframe
charpy_df =  pd.read_excel("Charpy_Impact_Test_Dataset.xlsx", header=1)
charpy_df
```

```python
# Check the data information for each column
charpy_df.info()
```

# 2. Visualizing the Distribution of Features in the Dataset <a name='2.-visualizing-the-distribution-of-features-in-the-dataset'></a>

```python
# Display the materials value counts
charpy_df['Material'].value_counts()
```

```python
# Standardize the Material column
charpy_df['Material'] = charpy_df['Material'].apply(
    lambda x: 'SA508' if pd.notna(x) and isinstance(x, str) and 'A508' in x.upper() 
    else ('SA533B' if pd.notna(x) and isinstance(x, str) and 'A533B' in x.upper() else x)
)
print("Material column after standardization:")
print(charpy_df['Material'].value_counts())
```

```python
# Grouping the material SS316
steel_mapping = {
    'SS316L': 'SS316',
    'SS316': 'SS316'
}
charpy_df['Material'] = charpy_df['Material'].replace(steel_mapping)
steel_counts = charpy_df['Material'].value_counts()
print(steel_counts)
```

## 2.1 Material Distribution <a name='2.1-material-distribution'></a>

```python
# Histogram of Material Distribution
steels_to_include = ['20MnMoNi55', 'A302B', 'SA508', 'SA533B', 'SS316']
filtered_df = charpy_df[charpy_df['Material'].isin(steels_to_include)]
steel_counts = filtered_df['Material'].value_counts()
desired_order = [ 'SA508','SA533B','20MnMoNi55','SS316', 'A302B']
steel_counts = steel_counts.reindex(desired_order)
colors = ['red','green','purple','blue','yellow']
plt.figure(figsize=(12, 6), dpi=300)
plt.bar(steel_counts.index, steel_counts.values, color=colors)
plt.xlabel('Material', fontsize=18,labelpad=8)
plt.ylabel('Count', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.show()
```

## 2.2 Specimen Dimensions <a name='2.2-specimen-dimensions'></a>

```python
# Histogram of Width (mm)
data = charpy_df['Width (mm)'].dropna()
counts, bins = np.histogram(data, bins='auto') 
midpoints = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(5, 4))
plt.bar(midpoints, counts, width=np.diff(bins), color='#00008B', edgecolor='black', align='center')
custom_ticks = [2.5, 5, 7.5, 10]
plt.xticks(custom_ticks, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Width (mm)', fontsize=14, labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram of Thickness (mm)
data = charpy_df['Thickness (mm)'].dropna()
counts, bins = np.histogram(data, bins='auto')
midpoints = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(5, 4))
plt.bar(midpoints, counts, width=np.diff(bins), color='#00008B', edgecolor='black', align='center')
custom_ticks = [2.5, 5, 7.5, 10]
plt.xticks(custom_ticks, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Thickness (mm)', fontsize=14,labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram of Length (mm)
data = charpy_df['Length (mm)'].dropna()
counts, bins = np.histogram(data, bins='auto')
midpoints = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(5, 4))
plt.bar(midpoints, counts, width=np.diff(bins), color='#00008B', edgecolor='black', align='center')
custom_ticks = np.linspace(data.min(), data.max(), 5)
plt.xticks(custom_ticks, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Length (mm)', fontsize=14, labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram of Notch Depth (mm)
data = charpy_df['Notch Depth (mm)'].dropna()
counts, bins = np.histogram(data, bins='auto')
midpoints = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(5, 4))
plt.bar(midpoints, counts, width=np.diff(bins), color='#00008B', edgecolor='black', align='center')
custom_ticks = [0.5, 1.0, 1.5, 2.0]
plt.xticks(custom_ticks, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Notch Depth (mm)', fontsize=14,labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram of Notch Angle (deg)
data = charpy_df['Notch Angle (deg)'].dropna()
counts, bins = np.histogram(data, bins='auto')
midpoints = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(5, 4))
plt.bar(midpoints, counts, width=np.diff(bins), color='#00008B', edgecolor='black', align='center')
custom_ticks = [30, 35, 40, 45]
plt.xticks(custom_ticks, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Notch Angle (deg)', fontsize=14,labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

## 2.3 Irradiation Conditions <a name='2.3-irradiation-conditions'></a>

```python
# Histogram of Irradiation Dose
charpy_df['Irradiation Dose (n/cm^2)'] = pd.to_numeric(charpy_df['Irradiation Dose (n/cm^2)'], errors='coerce')
irradiation_df = charpy_df.dropna(subset=['Irradiation Dose (n/cm^2)'])
irradiation_values = charpy_df['Irradiation Dose (n/cm^2)']
plt.figure(figsize=(5, 4))
plt.hist(irradiation_values, bins=30, color='#00008B', edgecolor='black')  # Adjust bins for better resolution
plt.xlabel(r'Irradiation Dose (n/cm$^2$)', fontsize=16,labelpad=8)
plt.ylabel('Count', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram of Irradiation Temperature
temperature_values = charpy_df['Irradiation Temperature (ºC)']
plt.figure(figsize=(5,4))
plt.hist(temperature_values, bins=20, color='#00008B', edgecolor='black')
plt.xlabel(r'Irradiation Temperature (°C)', fontsize=16,labelpad=8)
plt.ylabel('Count', fontsize=16)
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```

## 2.4 Charpy Properties <a name='2.4-charpy-properties'></a>

```python
# Histogram of Upper Shelf Energy
use_data = charpy_df['USE (J)'].dropna()
use_counts, use_bins = np.histogram(use_data, bins=20)
use_midpoints = (use_bins[:-1] + use_bins[1:]) / 2
plt.figure(figsize=(5, 4))
plt.bar(use_midpoints, use_counts, width=np.diff(use_bins), color='#00008B', align='center')
plt.xticks(np.linspace(min(use_data), max(use_data), 5), fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(r'Upper Shelf Energy (J)', fontsize=14, labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Histogram of Ductile-to-Brittle Transition Temperature
dbtt_data = charpy_df['DBTT (ºC)'].dropna()
dbtt_counts, dbtt_bins = np.histogram(dbtt_data, bins=20)
dbtt_midpoints = (dbtt_bins[:-1] + dbtt_bins[1:]) / 2
plt.figure(figsize=(5.2, 4))
plt.bar(dbtt_midpoints, dbtt_counts, width=np.diff(dbtt_bins), color='#00008B', align='center')
plt.xticks(np.linspace(min(dbtt_data), max(dbtt_data), 5), fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(r'Ductile-to-Brittle Transition Temperature (°C)', fontsize=14, labelpad=8)
plt.ylabel('Count', fontsize=14)
plt.grid()
plt.tight_layout()
plt.show()
```
